```python
import keras
from IPython.display import Image, HTML
json_file = open(r"C:\Users\ashwi\Downloads\model.json", 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = keras.models.model_from_json(loaded_model_json)
# load weights into new model
loaded_model.load_weights(r"C:\Users\ashwi\Downloads\model.h5")
```

```python
from tensorflow.keras.preprocessing import image
```

```python
import numpy as np
IMAGE_SIZE = (48,48)

img_path=r"C:\Users\ashwi\Downloads\face.jpg"
test_image=image.load_img(img_path,target_size= IMAGE_SIZE,color_mode='grayscale')
test_image=image.img_to_array(test_image)
print(test_image.shape)

img=image.load_img(img_path,target_size=IMAGE_SIZE)
img = np.expand_dims(img, axis=0)
print(img.shape)

result=loaded_model.predict(img)
print(result[0])
y_pred=np.argmax(result[0])
classes=['Angry','Disgust','Happy','Neutral','Sad','Surprise']
print('The person facial emotion is:',classes[y_pred])
```

```python
import numpy as np
import cv2
model = loaded_model
# prevents openCL usage and unnecessary logging messages
cv2.ocl.setUseOpenCL(False)

# dictionary which assigns each label an emotion (alphabetical order)
classes=['Angry','Disgust','Happy','Neutral','Sad','Surprise']

# start the webcam feed
cap = cv2.VideoCapture(0)
while True:
    # Find haar cascade to draw bounding box around face
    ret, frame = cap.read()
    if not ret:
        break
    facecasc = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    faces = facecasc.detectMultiScale(gray,scaleFactor=1.3, minNeighbors=5)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y-50), (x+w, y+h+10), (255, 0, 0), 2)
        roi_gray = gray[y:y + h, x:x + w]
        resized_img = np.expand_dims(cv2.resize(roi_gray, (48, 48)), 0)
        prediction = model.predict(resized_img)
        maxindex = int(np.argmax(prediction))
        cv2.putText(frame, classes[maxindex], (x+20, y-60), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)

    cv2.imshow('Video', cv2.resize(frame,(1920,1080),interpolation = cv2.INTER_CUBIC))
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
